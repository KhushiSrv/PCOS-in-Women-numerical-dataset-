```python
dataset.corr()['PCOS (Y/N)'].sort_values(ascending = False)
```

```python
dataset = dataset[['PCOS (Y/N)', ' Age (yrs)', 'Weight (Kg)','Height(Cm) ','Cycle(R/I)', 'Cycle length(days)', 
                   'Hip(inch)','Waist(inch)','Weight gain(Y/N)','Skin darkening (Y/N)','Hair loss(Y/N)','Pimples(Y/N)','Reg.Exercise(Y/N)']]
```

```python
dataset
```

```python
attribute = dataset.drop('PCOS (Y/N)', axis = 1).values
target = dataset['PCOS (Y/N)']
```

```python
attribute
```

```python
target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
attribute = scaler.fit_transform(dataset)
attribute
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(attribute, target , test_size=0.20, shuffle=False)
```

```python
print(X_train.shape,
Y_train.shape,
X_test.shape,
Y_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import  Dense
```

```python
Y_train = tf.keras.utils.to_categorical(Y_train, num_classes=2)
Y_test = tf.keras.utils.to_categorical(Y_test, num_classes=2)
```

```python
model = Sequential()

model.add(Dense(432, activation='relu',input_shape=(13,),))
model.add(Dense(10, input_dim=13, activation='relu'))
model.add(Dense(2,activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, batch_size=32, epochs=2, validation_split=0.1)
```

```python
model.summary()

test_loss, test_accuracy = model.evaluate(X_test, Y_test)
print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")
```

```python
pred =  model.predict(X_test)
```

```python
model.evaluate(X_test,pred)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svm = SVC(kernel='linear')
svm.fit(X_train, Y_train)
y_pred = svm.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```
